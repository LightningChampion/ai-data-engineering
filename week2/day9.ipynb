```python
import pandas as pd

data = {
    "product_name": ["laptop","mouse","desk","chair",None],
    "category": ["technology","technology","furniture",None,"furniture"],
    "price":[30000,500,None,2500,4000],
    "order_date": ["2026-06-01", "2026-06-02", "wrong-date", "2026-06-04",None]
}

df = pd.DataFrame(data)

df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.fillna("unknown")
```

```python
df["price"].fillna(0)
```

```python
df["category"].fillna(0)
```

```python
df["category"] = df["category"].fillna("unknown")
df
```

```python
df["product_name"]= df["product_name"].fillna("unknown")
df
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df
```

```python
df["order_date"].isnull().sum()
```

```python
df= df.dropna(subset=["order_date"])
df
```

```python
df = df.sort_values("price",ascending=False)
df
```

```python
df["category"].value_counts()
```

```python
df=df.rename(columns={"product_name":"product"})
df
```

```python
df["price"]=df["price"].astype(int)
df
```

```python
df.isnull().sum()
```

```python
df.to_csv("clean_products.csv", index=False)
```
